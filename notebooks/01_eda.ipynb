## Descripción del dataset

Dataset de transacciones reales de tarjetas de crédito europeas de septiembre de 2013.

- **284.807 transacciones** en total
- **492 son fraude** (0.17% del total)
- **Sin valores nulos**
- **Todas las variables son numéricas**

### Columnas

| Columna | Descripción |
|---|---|
| `Time` | Segundos transcurridos desde la primera transacción |
| `Amount` | Importe de la transacción en euros |
| `Class` | **0 = legítima, 1 = fraude** (variable objetivo) |
| `V1` - `V28` | Variables anonimizadas mediante PCA por privacidad |

> Las columnas V1-V28 son el resultado de transformar los datos reales del banco.
> No sabemos qué significan originalmente, pero son las más importantes para el modelo.

```python
# Contamos cuámtas transacciones hay de cada clase 
conteo = df["Class"].value_counts()
porcentaje = df["Class"].value_counts(normalize=True) * 100

print("Conteo absoluto:")
print(f" Legítimas(0): {conteo[0]:,}")
print(f" Fraudes(1): {conteo[1]:,}")

print("\nPorcentajes:")
print(f" Legítimas(0): {porcentaje[0]:.2f}%")
print(f" Fraudes(1): {porcentaje[1]:.2f}%")
```

```python
# Visualizamos del desbalance.

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Gráfico de barras
axes[0].bar(["legítima (0)", "Fraude (1)"], conteo.values, color=["steelblue", "crimson"])
axes[0].set_title("Conteo de transacciones por clase")
axes[0].set_ylabel("Número de transacciones")

# Gráfico de tarta
axes[1].pie(conteo.values, labels=["Legítima (0)", "Fraude (1)"],
            colors=["steelblue", "crimson"],
            autopct="%1.2f%%", startangle=90)
axes[1].set_title("Proporción de clases")

plt.tight_layout()
plt.savefig("../notebooks/images/class_distribution.png", dpi=150, bbox_inches="tight")
plt.show()
```

- El de barras muestra que la barra de fraude es prácticamente invisible. El desbalance es brutal. 
- El de tarta confirma el 0.17% vs 99.83%.

Esto justifica visualmente que necesitaremos SMOTE más adelante.

```python
# Separamos las transacciones legítimas y fraudulentas
legitimas = df[df["Class"] == 0]["Amount"]
fraudes = df[df["Class"] == 1]["Amount"]

print("Importes - Transacciones LEGÍTIMAS:")
print(f" Media: {legitimas.mean():.2f} €")
print(f" Mediana: {legitimas.median():.2f} €")
print(f" Máxima: {legitimas.max():.2f} €")

print("\nImportes - Transacciones FRAUDULENTAS:")
print(f" Media: {fraudes.mean():.2f} €")
print(f" Mediana: {fraudes.median():.2f} €")
print(f" Máxima: {fraudes.max():.2f} €")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Legítimas
axes[0].hist(legitimas, bins=50, color="steelblue", alpha=0.7)
axes[0].set_title("Importe — Transacciones legítimas")
axes[0].set_xlabel("Importe (€)")
axes[0].set_ylabel("Frecuencia")
axes[0].set_xlim(0, 500)  # Limitamos para ver mejor la distribución

# Fraudes
axes[1].hist(fraudes, bins=50, color="crimson", alpha=0.7)
axes[1].set_title("Importe — Transacciones fraudulentas")
axes[1].set_xlabel("Importe (€)")
axes[1].set_xlim(0, 500)

plt.tight_layout()
plt.savefig("../notebooks/images/amount_distribution.png", dpi=150, bbox_inches="tight")
plt.show()
```

Legítimas:
- Media 88€ pero mediana 22€ — hay algunas compras muy grandes (hasta 25.691€) que suben mucho la media, pero la mayoría son pequeñas.

Fraudulentas:
- Media 122€ pero mediana 9.25€ — la mediana es brutalmente baja. La mitad de los fraudes son de menos de 9.25€. Confirma lo que veíamos en el gráfico: los fraudulentos hacen muchas transacciones pequeñas.

*Transacciones legítimas.* la distribución es relativamente uniforme hasta 500€. La gente compra de todo, importes variados.  
*Transacciones fraudulentas.* Hay un pico enorme en importes pequeños (0-50€). Los fraudulentos tienden a hacer transacciones de poco importe, probablemente para no levantar sospechas o para probar si al tarjeta funciona antes de hacer cargos mayores.

Conclusión del análisis de Amount:  
La mediana de fraudes (9.25€) es menos de la mitad que la de legítimas (22€). El importe será una variable útil para el modelo, aunque no suficiente por sí sola.

```python
# Análisis de Time
# Time está en segundos, lo convertimos a horas para que sea más interpretable
df["Hour"] = (df["Time"] / 3600) % 24

legitimas_hora = df[df["Class"] == 0]["Hour"]
fraudes_hora = df[df["Class"] == 1]["Hour"]

print(f"El dataset cubre {df['Time'].max() / 3600:.1f} horas en total (~2 días)")
print(f"\nHora media transacciones legítimas: {legitimas_hora.mean():.1f}h")
print(f"Hora media transsaciones fraudulentas: {fraudes_hora.mean():.1f}h")
```

```python
# Visualización de Time
fig, axes= plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(legitimas_hora, bins=24, color="steelblue", alpha=0.7)
axes[0].set_title("Hora del día - Transacciones legítimas")
axes[0].set_xlabel("Hora del día")
axes[0].set_ylabel("Frecuencia")
axes[0].set_xticks(range(0, 25, 2))

axes[1].hist(fraudes_hora, bins=24, color="crimson", alpha=0.7)
axes[1].set_title("Hora del día - Transacciones fraudulentas")
axes[1].set_xlabel("Hora del día")
axes[1].set_xticks(range(0, 25,2))

plt.tight_layout()
plt.savefig("../notebooks/images/time_distribution.png", dpi=150, bbox_inches="tight")
plt.show()
```

*Transacciones legítimas* — patrón muy claro y lógico. Pocas transacciones de madrugada (2-7h), y mucha actividad durante el día (10-22h). Comportamiento humano normal.  
*Transacciones fraudulentas* — patrón completamente diferente. Hay fraudes a cualquier hora, incluso de madrugada. Los picos más llamativos son a las 2h y a las 12h. No siguen el patrón humano normal.

Conclusión del análisis de Time:  
Los fraudes ocurren de forma más uniforme a lo largo del día, incluyendo horas de madrugada donde apenas hay actividad legítima. La hora del día tiene potencial discriminante para el modelo.

```python
# Calculamos la correlación de cada variable con Class
correlaciones = df.corr()["Class"].drop("Class").sort_values(key=abs, ascending=False)

print("Top 10 variables más correlacionadas con frade:")
print(correlaciones.head(10).round(3))
```

```python
# Visualización
fig, ax = plt.subplots(figsize=(10, 6))

colores = ["crimson" if v < 0 else "steelblue" for v in correlaciones.head(15)]

ax.barh(correlaciones.head(15).index, correlaciones.head(15).values, color=colores)
ax.axvline(x=0, color="black", linewidth=0.8)
ax.set_title("Top 15 variables más correlacionadas con fraude")
ax.set_xlabel("Correlación con Class")

plt.tight_layout()
plt.savefig("../notebooks/images/correlaciones.png", dpi=150, bbox_inches="tight")
plt.show()
```

*Variables con correlación negativa (rojo)* — cuando su valor sube, la probabilidad de fraude baja. Las más importantes son V17 (-0.326), V14 (-0.303) y V12 (-0.261).  
*Variables con correlación positiva (azul)* — cuando su valor sube, la probabilidad de fraude sube. V11 (0.155) y V4 (0.133).


Conclusión del análisis de correlaciones:  
V17, V14 y V12 son las variables más discriminantes. El modelo las usará intensamente para detectar fraudes. El resto de variables V tienen correlaciones bajas — no son inútiles, pero aportan menos individualmente.
